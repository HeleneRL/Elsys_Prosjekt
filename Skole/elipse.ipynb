```python
A = [[1,2],[2,3]]


import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 100)

def x(theta):
    return np.cos(theta)
def y(theta):
    return np.sin(theta)

x_vector = [x(i) for i in theta]
y_vector = [y(i) for i in theta]
plt.plot(x_vector, y_vector)
plt.show()

z = np.array([np.matmul(A, [x(i), y(i)]) for i in theta])
plt.plot(z[:,0], z[:,1])
plt.show()
```
